```python
import matplotlib.pyplot as plt
from iminpaint.data.dataloader.data_utils import generate_mask
import numpy as np
```

```python
rows, cols = 5, 5
masks = [generate_mask((1, 8), (20, 50), (20, 40), (-0, np.pi)).numpy() for _ in range(rows * cols)]

fig, ax = plt.subplots(rows, cols, figsize=(cols * 10, rows * 10))
for i in range(rows):
    for j in range(cols):
        ax[i][j].imshow(masks[i * rows + j])
print(np.unique(masks[0]))
```
